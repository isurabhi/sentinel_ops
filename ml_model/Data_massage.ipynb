```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient("mongodb://localhost:27017")
db = client["sentinel_ops_db"]
collection = db["teams_freeze"]

# Query to find documents where 'device_city' is None or an empty string
query = {
    'device_city': {'$in': [None, ""]}
}

# Define the projection to select specific fields (1 to include, 0 to exclude)
projection = {"_id": 1, "device_name": 1, "device_city": 1}

# Execute the query and projection
filtered_documents = collection.find(query, projection)

# Loop through each document and update 'device_city' to 'Other' using 'device_name' in the query
for doc in filtered_documents:
    # Create a custom query using 'device_name' from the current document
    custom_query = {
        '_id': doc['_id'],
        'device_name': doc['device_name'],
        'device_city': {'$in': [None, ""]}
    }
    
    # Update operation
    update = {
        '$set': {
            'device_city': 'Other'
        }
    }
    
    # Perform the update
    collection.update_many(custom_query, update)
```
